```python
import pandas as pd
import numpy as np
import sklearn
```

```python
data = pd.read_csv('GIST_radiomicFeatures.csv')
```

```python
y = data['label']
variables = data.drop(['label','ID'],axis=1)
```

```python
data.var()
```

```python
from sklearn.feature_selection import VarianceThreshold
variance_tres = VarianceThreshold(0)
selection = variance_tres.fit(variables)
a = selection.get_support()
```

```python
a
```

```python
dropped_columns_df = [column for column in variables.columns 
          if column not in variables.columns[a]]
```

```python
dropped_columns_df
```

```python
high_variance = variables.drop(dropped_columns_df,axis=1)
```

```python
index_label = list()
```

```python
len_y = len(y)
```

```python
for i in range(0,len_y):
    if y[i] == 'GIST':
        index_label.append(1)
    else:
        index_label.append(0)    
```

```python
len(index_label)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
X_train, X_test, y_train, y_test = train_test_split(high_variance,index_label, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 1000, random_state = 0)
rf.fit(X_train, y_train)
```

```python
prediction = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction)
```
